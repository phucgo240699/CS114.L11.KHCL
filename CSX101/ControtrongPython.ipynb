**1. ĐẾM NODE LÁ**

```
class Node(object):
  def __init__(self, x):
    self.value = x
    self.l = None
    self.r = None
  
class Tree(object):
  def __init__(self):
    self.root = None
    self.count = 0

  def add(self, x):
    if self.root == None:
      self.root = Node(x)
    else:
      self._add(self.root, x)
  def _add(self, t, x):
    if t == None or t.value == x:
      pass
    else:
      if x < t.value:
        if t.l == None:
          t.l = Node(x)
        else:
          self._add(t.l, x)
      elif t.value < x:
        if t.r == None:
          t.r = Node(x)
        else:
          self._add(t.r, x)

  def LNR(self):
    if self.root != None:
      self._LNR(self.root)

  def _LNR(self, t):
    if t != None:
      self._LNR(t.l)
      if t.l == None and t.r == None:
        self.count += 1
      self._LNR(t.r)

a = Tree()

n = int(input())
while n != 0:
  a.add(n)
  n = int(input())

a.LNR()
print(a.count)
```

**2. CHIỀU CAO CÂY**

```

from collections import deque

class Node(object):
    def __init__(self, x):
        self.val = x
        self.l = None
        self.r = None

class Tree(object):
    def __init__(self):
        self.root = None

    def add(self, x):
        if self.root == None:
            self.root = Node(x)
        else:
            tmp = self.root
            while tmp != None:
                if tmp.val == x:
                    break
                elif x < tmp.val:
                    if tmp.l == None:
                        tmp.l = Node(x)
                        break
                    else:
                        tmp = tmp.l
                elif tmp.val < x:
                    if tmp.r == None:
                        tmp.r = Node(x)
                        break
                    else:
                        tmp = tmp.r


    def height(self, n):
      if n == None:
        return 0
      if self.height(n.l) >= self.height(n.r):
        return self.height(n.l) + 1
      else:
        return self.height(n.r) + 1


t = Tree()

st = None
arr = deque()
while st != '0':
  st = input()
  status = st.split()[0]
  if status == '3':
    break
  elif status == '2':
    a, b = st.split()[1], st.split()[2]
    if a in arr:
      arr.insert(arr.index(a) + 1, b)
    else:
      arr.appendleft(b)
  else:
    ID = st.split()[1]
    if status == '0':
      arr.appendleft(ID)
    elif status == '1':
      arr.append(ID)

for i in arr:
  t.add(int(i))

print(t.height(t.root))
```

**3. DUYỆT CÂY THEO CHIỀU RỘNG**

```

class Node(object):
    def __init__(self, x):
        self.val = x
        self.l = None
        self.r = None

class Tree(object):
    def __init__(self):
        self.root = None

    def add(self, x):
        if self.root == None:
            self.root = Node(x)
        else:
            tmp = self.root
            while tmp != None:
                if tmp.val == x:
                    break
                elif x < tmp.val:
                    if tmp.l == None:
                        tmp.l = Node(x)
                        break
                    else:
                        tmp = tmp.l
                elif tmp.val < x:
                    if tmp.r == None:
                        tmp.r = Node(x)
                        break
                    else:
                        tmp = tmp.r

    def BDF(self):
        if self.root != None: 
            q = [self.root]
            while q != []:
                tmp = q.pop(0)
                print(tmp.val, end = ' ')
                if tmp.l != None:
                    q.append(tmp.l)
                if tmp.r != None:
                    q.append(tmp.r)



n = input()
t = Tree()
while n != '0':
    t.add(int(n))
    n = input()

t.BDF()  
```
